```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = r"C:\Users\dorae\IdeaProjects\381_project_v2\solveStats_20Cubes.csv"  # Change to your CSV file path
df = pd.read_csv(file_path)

column_name = df.columns[1]
time = df[column_name]
plt.figure(figsize=(8, 6))
plt.hist(time, bins=20, edgecolor="black", alpha=0.7)
plt.xlabel("Time to Solve Cube (s)", fontsize=16)
plt.ylabel("Frequency", fontsize=16)
plt.title(f"Solving Time Distrib for MiniCubeSolver. {df.shape[0]} Data Points Collected.", fontsize=18);
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()

column_name = df.columns[2]
moves = df[column_name]
plt.figure(figsize=(8, 6))
plt.hist(moves, bins=20, edgecolor="black", alpha=0.7)
plt.xlabel("Minimum # of Moves to Solve Cube (s)", fontsize=16)
plt.ylabel("Frequency", fontsize=16)
plt.title(f"Min # of Moves Distrib for MiniCubeSolver. {df.shape[0]} Data Points Collected.", fontsize=18);
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
mean = np.mean(time)
median = np.median(time)
std_dev = np.std(time)

print(f"Mean time: {mean:.3f}")
print(f"Median time: {median:.3f}")
print(f"Standard Deviation time: {std_dev:.3f}")
print()

mean = np.mean(moves)
median = np.median(moves)
std_dev = np.std(moves)

print(f"Mean moves: {mean:.3f}")
print(f"Median moves: {median:.3f}")
print(f"Standard Deviation moves: {std_dev:.3f}")
```
