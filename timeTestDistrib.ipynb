```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = r"C:\Users\dorae\IdeaProjects\381_project_v2\timeTests.csv"  # Change to your CSV file path
df = pd.read_csv(file_path)

column_name = df.columns[1]
data = df[column_name]

plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, edgecolor="black", alpha=0.7)

plt.xlabel("Time to Solve Cube (s)")
plt.ylabel("Frequency")
plt.title(f"Solving Time Distribution for the MiniCubeSolver. {df.shape[0]} Data Points Collected.");

plt.show()
```

```python
mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)

print(f"Mean: {mean:.3f}")
print(f"Median: {median:.3f}")
print(f"Standard Deviation: {std_dev:.3f}")
```
